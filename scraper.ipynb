```python
# The output of get_match_dicts() is a list of dictionaries, i have you extract the match ID from each dict element and output a list of ID'scraper.ipynb
match_ids = {}

for item in matches:
    match_id = item['id']
    match_ids[match_id] = {
        'id': match_id,
        'home_team': item['homeTeam']['name'],
        'away_team': item['awayTeam']['name'],
        'round': item['roundInfo']['round'],
    }

match_ids
```

```python
# for each id in match_ids scrape the match data and save in a csv
for id in match_ids:
    match_stats = ss.scrape_team_match_stats(id)
    match_stats.to_csv(f"{match_ids[id]['home_team'].strip()}_{match_ids[id]['away_team'].strip()}_{match_ids[id]['id']}.csv", index=False)
```

```python
# Trying to loop over every match file and change the name to lower case and replace " " with "_"
import os

# Get folder
folder = r"epl_data_2425" # Path to folder
counter = 0

for filename in os.listdir(folder):
    if not os.path.isfile(os.path.join(folder, filename)):
        continue

    # Create new file name
    new_name = filename.lower().replace(" ", "_")

    if new_name != filename:
        old_path = os.path.join(folder, filename)
        new_path = os.path.join(folder, new_name)

    # Prevent overwriting if target already exists
    # if os.path.exists(new_path):
    #     print(f"Skipping {filename} -> {new_name} (target already exists)")
    #     continue
    replacements = {
        'brighton_&_hove_albion': 'brighton & hove albion',
        'aston_villa': 'aston villa',
        'crystal_palace': 'crystal palace',
        'manchester_united': 'manchester united',
        'leicester_city': 'leicester city',
        'manchester_city': 'manchester city',
        'newcastle_united': 'newcastle united',
        'ipswich_town': 'ipswich town',
        'west_hame': 'west ham',
        'nottingham_forest': 'nottingham'
    }

    try:
        os.rename(old_path, new_path)
    except:
        print(Exception)
    counter += counter

print(f"Renamed {counter} games")
```

```python
# Import json
import json
import os
# Loop through the 2425_data folder and flatten every game as row in a new dataframe

# Get folder
folder = "epl_data_2425"
all_rows = []                     # ← collect dictionaries here

# You probably want to load matches_data.json once outside the loop
with open('matches_data.json', 'r', encoding='utf-8') as f:
    matches = json.load(f)

for filename in os.listdir(folder):
    if not filename.endswith('.csv'):
        continue
        
    file_path = os.path.join(folder, filename)
    file_name = filename.removesuffix('.csv').lower()
    
    try:
        file_match_id = file_name.split('_',)[-1]
    except:
        print(f"Bad filename format: {filename}")
        continue
        
    df = pd.read_csv(file_path)
    
    row_dict = {}
    for _, row in df.iterrows():
        feature = (
            row["name"]
            .lower()
            .replace(" ", "_")
            .replace("%", "")
            .replace("(", "")
            .replace(")", "")
        )
        row_dict[f"{feature}_home"] = row["homeValue"]
        row_dict[f"{feature}_away"] = row["awayValue"]
    
    # Find match info
    found = False
    for m in matches:
        match_id = str(m['id'])
        if match_id == file_match_id:
            row_dict["home_h1_goals"]    = m['homeScore'].get('period1', 0)
            row_dict["home_h2_goals"]    = m['homeScore'].get('period2', 0)
            row_dict["home_normaltime"]  = m['homeScore'].get('normaltime', 0)
            row_dict["away_h1_goals"]    = m['awayScore'].get('period1', 0)
            row_dict["away_h2_goals"]    = m['awayScore'].get('period2', 0)
            row_dict["away_normaltime"]  = m['awayScore'].get('normaltime', 0)
            row_dict["game_id"]          = m['id']
            row_dict["game_winner"]      = m['winnerCode']
            row_dict["home_team"]  = m['homeTeam']['name'].lower().replace(' ', '_')
            row_dict["away_team"]  = m['awayTeam']['name'].lower().replace(' ', '_')
            row_dict['nameCode_home'] = m['homeTeam']['nameCode']
            row_dict['nameCode_away'] = m['awayTeam']['nameCode']
            found = True
            break
    
    all_rows.append(row_dict)

# ────────────────────────────────────────────────
# Create the final DataFrame **once**, after the loop
# ────────────────────────────────────────────────
season_df = pd.DataFrame(all_rows)
season_df = season_df.set_index('game_id')

# Optional: sort, clean types, etc.
season_df = season_df.sort_index()

print(season_df.shape)
season_df.head()
```
